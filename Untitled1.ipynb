```python
import scrapy
import requests
from bs4 import BeautifulSoup
from scrapy.crawler import CrawlerProcess
from phi_3_vision_mlx import load, generate, Phi3VProcessor
from serpapi import GoogleSearch
```

```python
# Replace 'your_model_path' with the actual path to Phi-3 Vision MLX
MODEL_PATH = "/Users/syedimamali/Phi-3-Vision-MLX/models/phi3_mini_128k"  # Example path

# Load Model & Processor with Correct Path
model = load()
processor = Phi3VProcessor(local_dir=MODEL_PATH)  #Pass the correct local directory
print("Model Loaded Successfully!")

# SERP API Key (Replace with your actual API key)
SERP_API_KEY = "078f5a140d9ab7ba96713f1a23042591dd9ccbd43e99f56163fd690574fd0102"
```

```python
def get_news_urls(query="latest AI news", num_results=20):
    """Fetches the latest AI news article URLs from Google News."""
    params = {
        "q": query,
        "api_key": SERP_API_KEY,
        "engine": "google_news"
    }
    
    response = requests.get("https://serpapi.com/search", params=params)
    news_data = response.json()
    
    # Extract the first 20 article URLs
    return [news["link"] for news in news_data.get("news_results", [])][:num_results]

# Fetch article URLs
article_urls = get_news_urls()
print(f"âœ… Found {len(article_urls)} articles.")
```

```python
from scrapy.crawler import CrawlerRunner
from twisted.internet import reactor

class NewsSpider(scrapy.Spider):
    name = "news_spider"
    
    def start_requests(self):
        """Fetch and scrape each article URL dynamically."""
        for url in article_urls:
            yield scrapy.Request(url=url, callback=self.parse_article)

    def parse_article(self, response):
        """Extract and summarize article content."""
        soup = BeautifulSoup(response.text, "html.parser")
        paragraphs = soup.find_all("p")
        article_text = " ".join([p.get_text() for p in paragraphs])

        if article_text:
            summary = self.summarize_text(article_text)
            print(f"\nðŸ”¹ Scraped: {response.url}")
            print("âœ… Summary:\n", summary)

    def summarize_text(self, text):
        """Summarizes an article using Phi-3 Vision MLX."""
        prompt = f"Summarize the following news article:\n{text}\n\nSummary:"
        tokens = processor.tokenizer(prompt, return_tensors="pt").input_ids
        output_tokens = generate(model, tokens, max_tokens=200)
        return processor.tokenizer.decode(output_tokens, skip_special_tokens=True)
```

```python
process = CrawlerProcess()
process.crawl(NewsSpider)
process.start()
```
